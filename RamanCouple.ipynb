```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import eig
```

```python
# parameters definitions
delta = 4.0 
omega = 3.5
gamma = 10.0
```

```python
single_hamiltonian  = np.array([[-delta/2, omega/2], [omega/2, delta/2]], dtype=complex)
double_hamiltonian = np.kron(single_hamiltonian, single_hamiltonian)
state_1 = np.array([0, 1], dtype=complex)
state_0 = np.array([1, 0], dtype=complex)
state_singlet = (np.kron(state_1, state_0) - np.kron(state_0, state_1)) / np.sqrt(2)
state_triplet_1 = np.kron(state_1, state_1)
state_triplet_0 = (np.kron(state_1, state_0) + np.kron(state_0, state_1)) / np.sqrt(2)
state_triplet_minus_1 = np.kron(state_0, state_0)
dissipation_term_singlet = 1j * gamma/2 * np.outer(state_singlet, state_singlet)
dissipation_term_triplet = 1j * gamma/2 * np.outer(state_triplet_0, state_triplet_0)
nh_hamiltonian = double_hamiltonian + dissipation_term_triplet
[eigenvalues, eigenvectors] = eig(nh_hamiltonian)
print(double_hamiltonian)
print(state_singlet)
print(state_triplet_1)
print(state_triplet_0)
print(state_triplet_minus_1)
print(dissipation_term_singlet)
print(dissipation_term_triplet)
print(nh_hamiltonian)
print(eigenvalues)
```
